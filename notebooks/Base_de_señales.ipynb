## **Listas en Python**

Una lista es una estructura que nos permite guardar varios valores juntos, en un orden específico. En Python, el primer elemento de una lista tiene el índice `0`, el segundo el índice `1`, y así sucesivamente.

```python
ags2 = [
    "Alex",  # índice 0
    "Hugo",  # índice 1
    "Erick", # índice 2
    "David"  # índice 3
]
```

```python
ags2
```

Podemos acceder a un elemento específico de la lista usando su índice:

```python
ags2[2]
```

## **Arreglos con NumPy**

NumPy es una biblioteca que nos permite trabajar con **arreglos** (arrays): colecciones de números sobre las que podemos hacer operaciones matemáticas de forma muy eficiente, algo que no es tan sencillo con listas normales.

```python
import numpy as np

np.random.random()
```

Cada vez que ejecutas esta celda obtienes un número distinto, entre 0 y 1:

```python
np.random.random()
```

Podemos crear un arreglo de números igualmente espaciados con `linspace`, y aplicarle una función matemática (como `sin`) a todos sus valores al mismo tiempo:

```python
x = np.linspace(0, 32000, 32000+1)
y = np.sin(x)
```

Grafiquemos el resultado:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30, 5))
plt.plot(x, y)
```

Al igual que con las listas, podemos tomar solo una parte ("rebanada" o *slice*) de un arreglo usando `[inicio:fin]`. Aquí tomamos los primeros 256 valores como una "ventana":


```python
k = 256

ventana_x = x[0:k]
ventana_y = y[0:k]
```

```python
plt.figure(figsize=(15, 5))
plt.plot(ventana_x, ventana_y)
```

# **Generando audio**

Una señal de audio digital no es más que un arreglo de números que representan la posición de una onda de sonido a lo largo del tiempo. Para generarla necesitamos definir:

- `fs`: la **frecuencia de muestreo**, es decir, cuántos valores (muestras) por segundo usamos para describir la onda.
- `duration`: la duración de la señal, en segundos.
- `f`: la frecuencia del tono que queremos generar, en Hz.

```python
import numpy as np
from IPython.display import Audio

fs = 44100    # frecuencia de muestreo (muestras por segundo)
duration = 2  # duración en segundos
f = 220       # frecuencia del tono, en Hz

t = np.linspace(0, duration, int(fs * duration))
wave = 0.5 * np.sin(2 * np.pi * f * t)
```

Podemos visualizar los primeros valores de nuestra onda:

```python
plt.plot(t[0:256], wave[0:256])
```

Y reproducir el audio directamente en el cuaderno:

```python
Audio(wave, rate=fs)
```

# **Funciones en Python**

Una función nos permite empaquetar un bloque de código para reutilizarlo cuantas veces queramos, con distintos valores de entrada (parámetros). Empecemos con un ejemplo sencillo, la función $f(x) = 2x + 1$:

```python
def funcion_lineal(x):
    return 2 * x + 1
```

```python
funcion_lineal(3)
```

```python
funcion_lineal(5)
```

Ahora apliquemos esta misma idea a lo que aprendimos sobre audio: empaquetemos la generación de un tono en una función, para poder crear distintos tonos fácilmente.

```python
def generar_tono(fs, f, duration):
    t = np.linspace(0, duration, int(fs * duration))
    wave = 0.5 * np.sin(2 * np.pi * f * t)

    return Audio(wave, rate=fs)
```

```python
generar_tono(44100, 440, 2)
```
